```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
```

# Load the data

```python
df = pd.read_csv("Gold Price (2013-2023).csv")
```

# Convert 'Date' to datetime and sort the dataframe by 'Date'

```python
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by='Date', ascending=True, inplace=True)
df.reset_index(drop=True, inplace=True)
```

# Extract year, month, and day from 'Date'


```python
df['Date_year'] = df['Date'].dt.year
df['Date_month'] = df['Date'].dt.month
df['Date_day'] = df['Date'].dt.day

print(df.info())
```

# Clean numerical columns by removing commas and converting to float


```python
columns = ["Price", "Open", "High", "Low"]
for col in columns:
    df[col] = df[col].replace({",": ""}, regex=True)
    df[col] = df[col].astype("float64")
```

# Replace 'K' with '000' and convert 'Vol.' to float


```python
df['Vol.'] = df['Vol.'].str.replace("K", "000", regex=True)
df['Vol.'] = df['Vol.'].astype("float64")
```

# Remove '%' sign from 'Change %' and convert to float


```python
df['Change %'] = df['Change %'].str.replace("%", "", regex=True)
df['Change %'] = df['Change %'].astype("float64")
```

# Check for and drop rows with NaN values in 'Vol.' and describe data


```python
print("Number of null values: ", df.isna().sum())
df = df.dropna(subset=['Vol.'])
```

```python
# Assuming 'df' is your DataFrame and it's already loaded and preprocessed
cols_to_shift = ["Open", "High", "Low", 'Vol.', "Change %"]

# Create a new DataFrame to store shifted values
df_shifted = df[['Date']].copy()

for col in cols_to_shift:
    df_shifted[f'Yesterday_{col}'] = df[col].shift(1)

# Merge the original DataFrame with the shifted DataFrame on 'Date'
df = df.merge(df_shifted, on='Date', how='left')

# Remove rows where the previous day's date does not match
df['Yesterday_Date'] = df['Date'].shift(1)
df = df[df['Date'] - pd.Timedelta(days=1) == df['Yesterday_Date']]

# Drop the temporary 'Yesterday_Date' column
df = df.drop(columns=['Yesterday_Date'])

# Reset index and drop any remaining rows with NaN values
df = df.reset_index(drop=True).dropna()

# Display the first few rows to verify the new columns
print(df.head())
```

# Remove outliers using the IQR method


```python
yesterday_columns = ["Yesterday_Open", "Yesterday_High", "Yesterday_Low", "Yesterday_Vol.", "Yesterday_Change %"]
for col in yesterday_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print(f"Number of outliers in {col}: ", len(df[(df[col] > upper_bound) | (df[col] < lower_bound)]))
    df = df[(df[col] <= upper_bound) & (df[col] >= lower_bound)]

print(df.head())
```

# Prepare data for modeling


```python
col = df.columns.drop(['Price', 'Date',"Open", "High", "Low", 'Vol.', "Change %"])
X = df[col]
y = df["Price"]
print(X.head())
print(y.head())
```

# Split the data into training and testing sets


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.30)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# create instance from LinearRegresiion 



```python
linreg= LinearRegression()
linreg.fit(X_train,y_train)
```

# Prediction and Plotting Actual vs Predicted values


```python
y_pred= linreg.predict(X_test) 
fig = plt.figure(figsize=(10,6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6)
sns.lineplot(x=y_test, y=y_test)
plt.xlabel('Actual Price', fontsize=14)
plt.ylabel('Predicted Price', fontsize=14)
plt.title('Actual vs Predicted Price', fontsize=17)
plt.show()
```

# Linear Regression Evaluation


```python
print("R2: ", r2_score(y_test, y_pred))
print("Mean Squared Error: ", np.sqrt(mean_squared_error(y_test, y_pred)))
print("Root Mean Squared Error: ", mean_squared_error(y_test, y_pred))
```
